```python
import yfinance as yf
import matplotlib.pyplot as plt

# 英伟达 vs AMD
tickers = ['NVDA', 'AMD']
data = yf.download(tickers, start='2020-01-01', end='2025-01-01')['Close']

print(data.head())
```

```python
normalized_data = data / data.iloc[0]

normalized_data.plot(figsize=(12, 6))
plt.title('Investment Return: NVDA vs AMD (2020-2025)')
plt.ylabel('Growth of $1 investment')
plt.grid(True)
plt.show()
```

```python
returns = data.pct_change()

volatility = returns.std()
print("年化波动率对比：")
print(volatility * (252**0.5)) 
```

```python
cumulative_returns = (1 + returns).cumprod()

rolling_max = data['NVDA'].cummax()
daily_drawdown = data['NVDA'] / rolling_max - 1
max_drawdown = daily_drawdown.min()

print(f"英伟达过去5年的最大跌幅（从最高点回落）: {max_drawdown:.2%}")

plt.figure(figsize=(12, 7))
plt.plot(cumulative_returns['NVDA'], label='NVIDIA (NVDA)', linewidth=2)
plt.plot(cumulative_returns['AMD'], label='AMD', alpha=0.7)
plt.title('Who Won the AI Chip War? (Cumulative Growth 2020-2025)', fontsize=15)
plt.xlabel('Year')
plt.ylabel('Growth of $1 Investment')
plt.legend()
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
data.to_csv('stock_prices_raw.csv')

cumulative_returns.to_csv('cumulative_returns_processed.csv')
print("数据已成功导出为 CSV 文件！")
```

Analysis and Business Insights: The Silicon Supremacy
1. The Great Divergence: Beyond Simple Stock Growth
While both NVIDIA (NVDA) and AMD operate within the high-stakes semiconductor industry, our data reveals a profound decoupling that began in early 2023. Historically, these two "chip giants" moved in relative tandem, yet the sudden explosion of Generative AI—triggered by the public release of ChatGPT—acted as a catalyst that propelled NVIDIA into a league of its own. By examining the cumulative returns, it becomes evident that an investor placing $1 into NVIDIA in 2020 would have seen their capital multiply far more aggressively than an identical investment in AMD. But what specifically drove this gap? My observation suggests it wasn't just about hardware sales; it was NVIDIA’s established "CUDA" software ecosystem that created a competitive moat, making it the indispensable backbone of the AI revolution.
2. Risk and Reward: Navigating the Volatility Storm
If the growth curve tells a story of triumph, the volatility metrics provide a sobering reality check for the cautious investor. High returns are rarely free of cost, and in NVIDIA’s case, the "price of admission" is a significantly higher annualized volatility compared to broader market indices. During our five-year window, we recorded several "dark moments" where the stock price experienced drawdowns exceeding 30%. This raises a critical question for portfolio managers: Is the potential for exponential growth worth the psychological stress of such sharp corrections? For NVIDIA, the volatility isn't merely "noise"—it reflects the market’s intense sensitivity to global supply chain shifts and AI regulatory news.
3. Synthesis: From Hardware Provider to AI Sovereign
Synthesizing these findings, we can conclude that NVIDIA has successfully transitioned from a gaming-focused GPU manufacturer to a sovereign entity of the AI era. The data confirms that while AMD remains a formidable competitor with strong fundamentals, NVIDIA’s ability to capture the lion’s share of the data center market has fundamentally altered its financial DNA. As we look forward, the primary challenge remains whether this trajectory is sustainable or if we are witnessing a "speculative bubble" reminiscent of the early internet era. For the intended user—be it a retail investor or a tech analyst—the takeaway is clear: NVIDIA is no longer just a stock; it is a high-beta proxy for the future of artificial intelligence itself.
